reemplazo de datos faltantes por la media


```python
# bore
COLUMNA=0
avg_bore=df['bore'].astype('float').mean(axis=COLUMNA)
avg_bore = round(avg_bore,4)
print("Average of bore:", avg_bore)
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

"horsepower": 2 datos perdidos, se reemplazar por la media
"peak-rpm": 2 datos perdidos, se reemplazar por la media

```python
# normalized-losses
COLUMNA=0
avg_normalized=df['normalized-losses'].astype('float').mean(axis=COLUMNA)
avg_normalized = round(avg_normalized,4)
print("Average normalized-losses:", avg_normalized)
df["normalized-losses"].replace(np.nan, avg_normalized, inplace=True)
```

```python
# stroke
COLUMNA=0
avg_stroke=df['stroke'].astype('float').mean(axis=COLUMNA)
avg_stroke = round(avg_stroke,4)
print("Average stroke:", avg_stroke)
df["stroke"].replace(np.nan, avg_stroke, inplace=True)
```

```python
# horsepower
COLUMNA=0
avg_horsepower=df['horsepower'].astype('float').mean(axis=COLUMNA)
avg_horsepower = round(avg_horsepower,4)
print("Average horsepower:", avg_horsepower)
df["horsepower"].replace(np.nan, avg_horsepower, inplace=True)
```

```python
# peak-rpm
COLUMNA=0
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=COLUMNA)
avg_peakrpm = round(avg_peakrpm,4)
print("Average peak-rpm:", avg_peakrpm)
df["peak-rpm"].replace(np.nan,avg_peakrpm, inplace=True)
```

```python
# Convirtiendo mpg a km : (235 dividido por mpg)
df['city-km'] = round(235/df["city-mpg"],3)

# mostramos datos
df[['city-mpg','city-km']].head()
```

Ejercicio , transformar la variables predictora "highway-mpg" a kilometros en la nueva variable predictora "highway-km"

```python
# Convirtiendo mpg a km : (235 dividido por mpg)
df['highway-km'] = round(235/df["highway-mpg"],3)

# mostramos datos
df[['highway-mpg','highway-km']].head()
```

```python
# replace (original value) by (original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['width'].max()
# mostramos datos
df[['length','width','height']].head()
```

Ejercicio , normalizar la variable predictora "height"

```python
# replace (original value) by (original value)/(maximum value)
df['height'] = df['height']/df['height'].max()
df[['height']].head()
```

```python
# Genera puntos de corte 
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_names = ['Bajo', 'Medio', 'Alto']
```

```python
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head()
```
